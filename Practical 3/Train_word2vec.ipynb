```python
import string
def read_words(filename):
    words = []
    translator = str.maketrans('', '', string.punctuation)
    scount = 0
    with open(filename) as f:
        for s in f:
            scount += 1
            clean_s = s.translate(translator).lower()
            words.append(clean_s.split())
    return words
```

```python
data = read_words('europarl-training.en')
```

```python
import gensim

model = gensim.models.Word2Vec(data, workers=3, sg=1, size = 300 )
```

```python
model.wv.most_similar('money')
```

```python
model.save('gensim_skipgram_model')
```
